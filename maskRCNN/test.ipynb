```python
import torch
from utils.dataset import StrawberryDataset
import torchvision.models as models
from torchvision.models.detection import maskrcnn_resnet50_fpn
from utils.masks_bb import masks_bb
import matplotlib.pyplot as plt
```

```python
model = maskrcnn_resnet50_fpn(progress=True, pretrained_backbone=True, num_classes=2, box_detections_per_img=12)
model.load_state_dict(torch.load('strawberry_model.pth'))
model.eval()
```

```python
masks_test, boxes_test = masks_bb('data/test/masks/')
```

```python
dataset = StrawberryDataset('data/test/Images/', 'data/test/masks/', boxes_test, masks_test)
```

```python
def collate_fn(batch):
    return tuple(zip(*batch))

dataloader = torch.utils.data.DataLoader(dataset, batch_size=4, shuffle=True, collate_fn=collate_fn)
```

```python
batch = next(iter(dataloader))
image, target = batch
output = model(image)
# plt.imshow(image[0].permute(1, 2, 0))
# plt.show()

dataset.show_image(image[0], output[0]['masks'], output[0]['boxes'])
```
